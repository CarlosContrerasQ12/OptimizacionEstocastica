```julia
using LinearAlgebra
```

```julia
function p(j,s,a)
    if s==1
        return P1[a,j]
    elseif s==2
        return P2[a,j]
    elseif s==3
        return P3[a,j]
    elseif s==4
        return P4[a,j]
    else
        return 0.0
    end
end    

function construir_matriz(a,S)
    A=zeros(length(S),length(S))
    for i in 1:length(S)
        for j in 1:length(S)
            A[j,i]=p(S[i],S[j],a[j])
        end
    end
    return A
end

function p_rec(s)
    if s==1
        return 10.0
    elseif s==2
        return 5.0
    elseif s==3
        return -1.0
    else s==4
        return -10.0
    end
end

function r1(sig,s,a)
    reca=p_rec(sig)
    if a==1
        reca=reca-1
    elseif a==2
        reca=reca-4
    elseif a==3
        reca=reca-10
    end
    return reca
end 
```

```julia
function iteracion_descontada(λ,d0p,rec,S,A)
    d0=deepcopy(d0p)
    d1=deepcopy(d0p)
    seguir=true
    while seguir
        Pd=construir_matriz(d1,S)
        Q=I(length(S))-λ*Pd
        r=construir_r(rec,d1,S,A)
        v=Q\r
        pnew=deepcopy(d1)
        Pdv=Pd*v
        for (i,s) in enumerate(S)
            val=[r_exp(s,a,rec,S,A)+λ*Pdv[i] for a in A[i]]
            pnew[i]=A[i][argmax(val)]
        end
        d0=deepcopy(d1)
        d1=deepcopy(pnew)
        if d1==d0
            seguir=false
        end
    end
    return d1
end
```

```julia
using QuantEcon

chain = DiscreteMarkovChain(construir_matriz([2,2,2,3],1:4)) 
stationary_distribution(chain)

function calcular_MRP(X,a,rec,S,A)
    suma=0.0
    l=0.0
    for (i,x) in enumerate(X[1:end-1])
        suma+=r_exp(x,a[x],rec,S,A)
        l+=1.0
    end
    return suma/l    
end

function calcular_esperanza(a,rec,S,A)
    suma=0.0
    l=0.0
    chain=QuantEcon.MarkovChain(construir_matriz(a,1:4))
    for i in 1:10000
        X = simulate(chain, 10_000)
        suma+=calcular_MRP(X,a,rec,S,A)
        l+=1
    end
    return suma/l
end

using PyPlot


chain=QuantEcon.MarkovChain(construir_matriz([1,1,2,2],1:4))
X = simulate(chain, 10_000)
y=(cumsum(X.==1)) ./ (1:length(X))
plot(y)
y[end]

calcular_esperanza([2,2,3,3],r1,S,A)
```
